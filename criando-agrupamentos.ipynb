## Criando faixas de valor

```python
dados_aux = pd.read_csv('dados/aluguel.csv', sep = ';')
dados_aux.head(10)
```

```python
# 1 e 2
# 3 e 4
# 5 e 6
# 7 ou mais
classes = [0, 2, 4, 6, 100]
```

```python
quartos = pd.cut(dados.Quartos, classes)
quartos
```

```python
pd.value_counts(quartos)
```

```python
labels = ['1 ou 2 quartos', '3 ou 4 quartos', '5 ou 6 quartos', '7 quartos ou mais']
```

```python
quartos = pd.cut(dados.Quartos, classes, labels = labels)
```

```python
pd.value_counts(quartos)
```

```python
quartos = pd.cut(dados.Quartos, classes, labels = labels, include_lowest = True)
```

```python
pd.value_counts(quartos)
```

##### Nota-se que existem imóveis com 0 quartos 11250 -> 11409, ou seja, 159 imóveis com 0 quartos
