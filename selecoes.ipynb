### Seleção de imóveis do tipo 'Apartamento'

```python
import pandas as pd
```

```python
dados = pd.read_csv('dados/aluguel_residencial.csv', sep = ';')
```

```python
apt = ['Apartamento']
```

```python
selecao1 = dados['Tipo'].isin(apt)
selecao1
```

```python
dados_apartamento = dados[selecao1]
dados_apartamento
```

```python
dados_apartamento.index = range(dados_apartamento.shape[0])
dados_apartamento
```

```python
dados_apartamento.to_csv('dados/aluguel_apartamento.csv', sep = ';', index = False)
```

```python
dados_apartamento_2 = pd.read_csv('dados/aluguel_apartamento.csv', sep = ';')
dados_apartamento_2
```

### Seleção de imóveis dos tipos 'Casa', 'Casa de Condomínio' e 'Casa de Vila'

```python
casas = ['Casa', 'Casa de Condomínio', 'Casa de Vila']
```

```python
selecao2 = dados['Tipo'].isin(casas)
selecao2
```

```python
dados_casas = dados[selecao2]
dados_casas 
```

```python
dados_casas.index = range(dados_casas.shape[0])
dados_casas
```

```python
dados_casas.to_csv('dados/aluguel_casas.csv', sep = ';', index = False)
```

```python
dados_casas_2 = pd.read_csv('dados/aluguel_casas.csv', sep = ';')
dados_casas_2
```

### Seleção de imóveis com área entre 60 e 100 metros quadrados (incluindo os limites)

```python
dados['Area'].dtype
```

```python
selecao3 = (dados['Area'] >= 60) & (dados['Area'] <= 100)
selecao3
```

```python
dados_area = dados[selecao3]
dados_area 
```

```python
dados_area.index = range(dados_area.shape[0])
dados_area
```

```python
dados_area.to_csv('dados/aluguel_area.csv', sep = ';', index = False)
```

```python
dados_area_2 = pd.read_csv('dados/aluguel_area.csv', sep = ';')
dados_area_2
```

### Seleção de imóveis que tenham pelo menos 4 quartos e aluguel menor que R$ 2.000,00

```python
selecao4 = (dados['Quartos'] >= 4) & (dados['Valor'] < 2000)
selecao4
```

```python
dados_quartos = dados[selecao4]
dados_quartos
```

```python
dados_quartos.index = range(dados_quartos.shape[0])
dados_quartos
```

```python
dados_quartos.to_csv('dados/aluguel_quartos.csv', sep = ';', index = False)
```

```python
dados_quartos_2 = pd.read_csv('dados/aluguel_quartos.csv', sep = ';')
dados_quartos_2
```

### Conclusões

```python
n1 = dados[selecao1].shape[0]
n2 = dados[selecao2].shape[0]
n3 = dados[selecao3].shape[0]
n4 = dados[selecao4].shape[0]
```

```python
print("Nº de imóveis classificados com tipo 'Apartamento' -> {}".format(n1))
print("Nº de imóveis classificados com tipos 'Casa', 'Casa de Condomínio' e 'Casa de Vila' -> {}".format(n2))
print("Nº de imóveis com área entre 60 e 100 metros quadrados (incluindo os limites) -> {}".format(n3))
print("Nº de imóveis que tenham pelo menos 4 quartos e aluguel menor que R$ 2.000,00 -> {}".format(n4))
```
